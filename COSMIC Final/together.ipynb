```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

EvpTrain20 = pd.read_csv("/Users/soumilhooda/Desktop/COSMIC Final/profile-evp20-train.ascii",delim_whitespace=' ')
EvpTest20 = pd.read_csv("/Users/soumilhooda/Desktop/COSMIC Final/profile-evp20-test.ascii",delim_whitespace=' ')
EvpTest21 = pd.read_csv("/Users/soumilhooda/Desktop/COSMIC Final/profile-evp21-test.ascii",delim_whitespace=' ')
TmpTrain20 = pd.read_csv("/Users/soumilhooda/Desktop/COSMIC Final/profile-tmp20-train.ascii",delim_whitespace=' ')
TmpTest20 = pd.read_csv("/Users/soumilhooda/Desktop/COSMIC Final/profile-tmp20-test.ascii",delim_whitespace=' ')
TmpTest21 = pd.read_csv("/Users/soumilhooda/Desktop/COSMIC Final/profile-tmp21-test.ascii",delim_whitespace=' ')
PrsTrain20 = pd.read_csv("/Users/soumilhooda/Desktop/COSMIC Final/profile-prs20-train.ascii",delim_whitespace=' ')
PrsTest20 = pd.read_csv("/Users/soumilhooda/Desktop/COSMIC Final/profile-prs20-test.ascii",delim_whitespace=' ')
PrsTest21 = pd.read_csv("/Users/soumilhooda/Desktop/COSMIC Final/profile-prs21-test.ascii",delim_whitespace=' ')

features = ['HGT', 'MIN-A', 'MEAN-A', 'MAX-A', 'STD-A', 'MIN-P', 'MEAN-P', 'MAX-P', 'STD-P', 'BIAS', 'RMSE', 'STD', 'CORR', 'COUNT']

EvpTrain20.columns = features
EvpTest20.columns = features
EvpTest21.columns = features
TmpTrain20.columns = features
TmpTest20.columns = features
TmpTest21.columns = features
PrsTrain20.columns = features
PrsTest20.columns = features
PrsTest21.columns = features
```

# PLOTS

```python
titledic = {
    'fontsize' : 20,
    'fontweight' : 5,
}
axesdic = {
    'fontsize' : 15,
    'fontweight' : 4,
}
```

```python
height = []
for i in range(0,16000, 500):
    height.append(i)


heightlabel = []
for i in range(0,16000, 500):
    heightlabel.append(i)

count21 = [400000, 500000, 600000, 700000, 800000, 900000, 1000000]
```

```python
from matplotlib import rc, rcParams
rc('mathtext', default='regular')
rcParams["figure.figsize"] = (25,15)
rcParams['axes.linewidth'] = 2

fig = plt.figure()
ax = fig.add_subplot(131)
ax.set_title('Humidity RMSE',fontdict=titledic, pad=15)
ax.plot(EvpTrain20['RMSE'], EvpTrain20['HGT'], '-', label = 'RMSE Train ', linewidth=2)
ax.plot(EvpTest20['RMSE'], EvpTest20['HGT'], '-', label = 'RMSE Test ', linewidth=2)
ax.plot(EvpTest21['RMSE'], EvpTest21['HGT'], '-', label = 'RMSE Unseen ', linewidth=2)

ax.legend(loc=0)
# ax.grid()
ax.set_xlabel("Humidity", fontdict=axesdic)
ax.set_ylabel(r"Height (Metres)", fontdict=axesdic)
ax.set_yticks(height, labels=heightlabel, minor=True)
ax.set_ylim(0,16000)
ax.set_xlim(0,3)


ax = fig.add_subplot(132)
ax.set_title('Temperature RMSE',fontdict=titledic, pad=15)
ax.plot(TmpTrain20['RMSE'], TmpTrain20['HGT'], '-', label = 'RMSE Train ', linewidth=2)
ax.plot(TmpTest20['RMSE'], TmpTest20['HGT'], '-', label = 'RMSE Test ', linewidth=2)
ax.plot(TmpTest21['RMSE'], TmpTest21['HGT'], '-', label = 'RMSE Unseen ', linewidth=2)

ax.legend(loc=0)
# ax.grid()
ax.set_xlabel("Temperature", fontdict=axesdic)
ax.set_ylabel(r"Height (Metres)", fontdict=axesdic)
ax.set_yticks(height, labels=heightlabel, minor=True)
ax.set_ylim(0,16000)
ax.set_xlim(0,3)


ax = fig.add_subplot(133)
ax.set_title('Pressure RMSE',fontdict=titledic, pad=15)
ax.plot(PrsTrain20['RMSE'], PrsTrain20['HGT'], '-', label = 'RMSE Train ', linewidth=2)
ax.plot(PrsTest20['RMSE'], PrsTest20['HGT'], '-', label = 'RMSE Test ', linewidth=2)
ax.plot(PrsTest21['RMSE'], PrsTest21['HGT'], '-', label = 'RMSE Unseen ', linewidth=2)

ax.legend(loc=0)
# ax.grid()
ax.set_xlabel("Pressure", fontdict=axesdic)
ax.set_ylabel(r"Height (Metres)", fontdict=axesdic)
ax.set_yticks(height, labels=heightlabel, minor=True)
ax.set_ylim(0,16000)
ax.set_xlim(0,3)



plt.show()
```
