```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

Train20 = pd.read_csv("/Users/soumilhooda/Desktop/COSMIC Final/profile-tmp20-train.ascii",delim_whitespace=' ')
Test20 = pd.read_csv("/Users/soumilhooda/Desktop/COSMIC Final/profile-tmp20-test.ascii",delim_whitespace=' ')
Test21 = pd.read_csv("/Users/soumilhooda/Desktop/COSMIC Final/profile-tmp21-test.ascii",delim_whitespace=' ')

features = ['HGT', 'MIN-A', 'MEAN-A', 'MAX-A', 'STD-A', 'MIN-P', 'MEAN-P', 'MAX-P', 'STD-P', 'BIAS', 'RMSE', 'STD', 'CORR', 'COUNT']

Train20.columns = features
Test20.columns = features
Test21.columns = features
```

# TEMPERATURE PLOTS

```python
titledic = {
    'fontsize' : 20,
    'fontweight' : 5,
}
axesdic = {
    'fontsize' : 15,
    'fontweight' : 4,
}
```

```python
height = []
for i in range(0,16000, 500):
    height.append(i)


heightlabel = []
for i in range(0,16000, 500):
    heightlabel.append(i)

count21 = [400000, 500000, 600000, 700000, 800000, 900000, 1000000]
```

```python

from matplotlib import rc, rcParams
rc('mathtext', default='regular')
rcParams["figure.figsize"] = (25,15)
rcParams['axes.linewidth'] = 2

fig = plt.figure()
ax = fig.add_subplot(131)
ax.set_title('Training Data', fontdict=titledic, pad=15)
ax.plot(Train20['MIN-A'], Train20['HGT'], '-', label = 'Minimum', linewidth=2)
ax.plot(Train20['MEAN-A'], Train20['HGT'], '-', label = 'Mean', linewidth=2)
ax.plot(Train20['MAX-A'], Train20['HGT'], '-', label = 'Maximum', linewidth=2)
ax2 = ax.twiny()
ax2.plot(Train20['COUNT'], Train20['HGT'], '-r', label = 'Sample Frequuency', linewidth=2)
# ax2.ticklabel_format(style='sci',scilimits=(-3,4), axes=X)
ax.legend(loc=0)
ax2.legend(loc=2)
# ax.grid()
ax.set_xlabel("Temperature", fontdict=axesdic)
ax.set_ylabel(r"Height (Metres)", fontdict=axesdic)
ax2.set_xlabel(r"Sample Frequency", fontdict=axesdic, labelpad=10)

ax.set_yticks(height, labels=heightlabel, minor=True)
# ax.tick_params(axis='both', pad=4)
ax.set_ylim(0,16000)
ax.set_xlim(150,350)


ax = fig.add_subplot(132)
ax.set_title('Testing Data', fontdict=titledic, pad=10)
ax.plot(Test20['MIN-A'], Test20['HGT'], '-', label = 'Minimum', linewidth=2)
ax.plot(Test20['MEAN-A'], Test20['HGT'], '-', label = 'Mean', linewidth=2)
ax.plot(Test20['MAX-A'], Test20['HGT'], '-', label = 'Maximum', linewidth=2)
ax2 = ax.twiny()
ax2.plot(Test20['COUNT'], Test20['HGT'], '-r', label = 'Sample Frequuency', linewidth=2)
# ax2.ticklabel_format(style='sci',scilimits=(-3,4), axes=X)
ax.legend(loc=0)
ax2.legend(loc=2)
# ax.grid()
ax.set_xlabel("Temperature", fontdict=axesdic)
ax.set_ylabel(r"Height (Metres)", fontdict=axesdic)
ax2.set_xlabel(r"Sample Frequency", fontdict=axesdic, labelpad=10)

ax.set_yticks(height, labels=heightlabel, minor=True)
# ax.tick_params(axis='both', pad=4)
ax.set_ylim(0,16000)
ax.set_xlim(150,350)


ax = fig.add_subplot(133)
ax.set_title('Unseen Data', fontdict=titledic, pad=10)
ax.plot(Test21['MIN-A'], Test21['HGT'], '-', label = 'Minimum', linewidth=2)
ax.plot(Test21['MEAN-A'], Test21['HGT'], '-', label = 'Mean', linewidth=2)
ax.plot(Test21['MAX-A'], Test21['HGT'], '-', label = 'Maximum', linewidth=2)
ax2 = ax.twiny()
ax2.plot(Test21['COUNT'], Test21['HGT'], '-r', label = 'Sample Frequuency', linewidth=2)
# ax2.ticklabel_format(style='sci',scilimits=(-3,4), axes=X)
ax.legend(loc=0)
ax2.legend(loc=2)
# ax.grid()
ax.set_xlabel("Temperature", fontdict=axesdic)
ax.set_ylabel(r"Height (Metres)", fontdict=axesdic)
ax2.set_xlabel(r"Sample Frequency", fontdict=axesdic, labelpad=10)

ax.set_yticks(height, labels=heightlabel, minor=True)
ax2.set_xticks(count21, labels=count21)
# ax.tick_params(axis='both', pad=4)
ax.set_ylim(0,16000)
ax.set_xlim(150,350)


plt.show()
```
